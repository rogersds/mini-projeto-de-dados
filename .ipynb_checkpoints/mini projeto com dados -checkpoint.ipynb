### O que queremos saber/fazer?

1. Valor Total da Folha Salarial -> Qual foi o gasto total com salários de funcionários pela empresa? <br>
    Sugestão: calcule o salário total de cada funcionário, salário + benefícios + impostos, depois some todos os salários
    
    
2. Qual foi o faturamento da empresa?<br>
    Sugestão: calcule o faturamento total de cada serviço e depois some o faturamento de todos
    
    
3. Qual o % de funcionários que já fechou algum contrato?<br>
    Sugestão: na base de serviços temos o funcionário que fechou cada serviço. Mas nem todos os funcionários que a empresa tem já fecharam algum serviço.<br>
    . Na base de funcionários temos uma lista com todos os funcionários<br>
    . Queremos calcular Qtde_Funcionarios_Fecharam_Serviço / Qtde_Funcionários_Totais<br>
    . Para calcular a qtde de funcionários que fecharam algum serviço, use a base de serviços e conte quantos funcionários tem ali. Mas lembre-se, cada funcionário só pode ser contado uma única vez.<br><br>
    Dica: se você aplicar o método .unique() em uma variável que é apenas 1 coluna de um dataframe, ele vai excluir todos os valores duplicados daquela coluna.<br>
    Ex: unicos_colunaA = dataframe['colunaA'].unique() te dá como resposta uma lista com todos os itens da colunaA aparecendo uma única vez. Todos os valores repetidos da colunaA são excluidos da variável unicos_colunaA 
    
    
4. Calcule o total de contratos que cada área da empresa já fechou


5. Calcule o total de funcionários por área


6. Qual o ticket médio mensal (faturamento médio mensal) dos contratos?<br>
    Dica: .mean() calcula a média -> exemplo: media_colunaA = dataframe['colunaA'].mean()

Obs: Lembrando as opções mais usuais de encoding:<br>
encoding='latin1', encoding='ISO-8859-1', encoding='utf-8' ou então encoding='cp1252'

Observação Importante: Se o seu código der um erro na hora de importar os arquivos:<br>
- CadastroClientes.csv
- CadastroFuncionarios.csv

Use separador ";" (ponto e vírgula) para resolver

```python
import pandas as pd
# no csv o separador dele é a ',', se tiver virgula em celulas ele vai dar erro
# segue abaxi oa forma de tratar esse erro
# ex: cadastro_clientes_df = pd.read_csv('CadastroClientes.csv', sep=';', decimal = '.')

funcionarios_df = pd.read_csv('CadastroFuncionarios.csv', sep=';', decimal=',')
clientes_df = pd.read_csv('CadastroClientes.csv', sep=';', decimal=',')
servicos_df = pd.read_excel('BaseServiçosPrestados.xlsx')

#retirar colunas Estado Civil e Cargo da tabela de funcionarios
funcionarios_df = funcionarios_df.drop(['Estado Civil', 'Cargo'], axis=1)

display(funcionarios_df)
display(clientes_df)
display(servicos_df)
```

## Valor Total da Folha Salarial -> Qual foi o gasto total com salários de funcionários pela empresa?
Sugestão: calcule o salário total de cada funcionário, salário + benefícios + impostos, depois some todos os salários

```python
# criando uma nova coluna com o sario total somando as colunas Salario Base	Impostos	Beneficios	VT	VR
funcionarios_df['Salario Total'] = funcionarios_df['Salario Base'] + funcionarios_df['Impostos'] + funcionarios_df['Beneficios'] + funcionarios_df['VT'] + funcionarios_df['VR']
print('Total da Folha Salarial Mensal é de R${:,}'.format(funcionarios_df['Salario Total'].sum()))
```

## 2. Qual foi o faturamento da empresa?<br>
    Sugestão: calcule o faturamento total de cada serviço e depois some o faturamento de todos

```python
# usando .merge para criar uma tabela auxiliar 
# fazendo passo a passo pra entender 
# juntando todas as tabelas
# faturamentos_df = servicos_df.merge(clientes_df)
# juntando apenas as colunas que quero dasd tabelas 
#faturamentos_df = servicos_df[['ID Cliente', 'Tempo Total de Contrato (Meses)']].merge(clientes_df[['ID Cliente' ,'Valor Contrato Mensal']], on='ID Cliente')
# on= vai pegar uma coluna com os mesmos dados "relacionais"de cada tabela para fazer a juçao 
# primeiro vc coloca a tabela que vai receber as info
# e dentro do .merge()vc coloca a tabela que vai dar as info

faturamentos_df = servicos_df[['ID Cliente', 'Tempo Total de Contrato (Meses)']].merge(clientes_df[['ID Cliente' ,'Valor Contrato Mensal']], on='ID Cliente')
faturamentos_df['Faturamento Total'] = faturamentos_df['Tempo Total de Contrato (Meses)'] * faturamentos_df['Valor Contrato Mensal']

print('Faturamento Total: R${:,}'.format(sum(faturamentos_df['Faturamento Total'])))
```

## 3. Qual o % de funcionários que já fechou algum contrato?

Dica: se você aplicar o método .unique() em uma variável que é apenas 1 coluna de um dataframe, ele vai excluir todos os valores duplicados daquela coluna.
Ex: unicos_colunaA = dataframe['colunaA'].unique() te dá como resposta uma lista com todos os itens da colunaA aparecendo uma única vez. Todos os valores repetidos da colunaA são excluidos da variável unicos_colunaA

```python
# usando o len para pegar o tamanho da coluna
# usando .unique() para retirar duplicatas 

Qtde_Funcionarios_Fecharam_Serviço = len(servicos_df['ID Funcionário'].unique())
Qtde_Funcionários_Totais = len(funcionarios_df['ID Funcionário'].unique())
#print(servivos_prestado)
#print(Qtde_Funcionários_Totais)

qtde_per_fecharam_pedido = Qtde_Funcionarios_Fecharam_Serviço / Qtde_Funcionários_Totais
print(' O % de funcionários que já fechou algum contrato foi de {:.2%}'.format(qtde_per_fecharam_pedido))
```

## Calcule o total de contratos que cada área da empresa já fechou

```python
# usando .merge()
# primeiro vc coloca a tabela que vai receber as info
# e dentro do .merge()vc coloca a tabela que vai dar as info
# on= vai pegar uma coluna com os mesmos dados "relacionais"de cada tabela para fazer a juçao 
#Usaremos o método .value_counts() para contar quantas vezes cada valor do Dataframe aparece

contratos_area_df = servicos_df[['ID Funcionário']].merge(funcionarios_df[['ID Funcionário', 'Area']], on='ID Funcionário')
contratos_area_qtde = contratos_area_df['Area'].value_counts()
print(contratos_area_qtde)

contratos_area_qtde.plot(kind='bar', figsize=(8, 5))
```

## Calcule o total de funcionários por área

```python
total_funcionarios = funcionarios_df['Area'].value_counts()

#  Usaremos o método .plot() para exibir um gráfico
total_funcionarios.plot(kind='bar', figsize=(8, 5))
print(total_funcionarios)

```

## Qual o ticket médio mensal (faturamento médio mensal) dos contratos?

```python
ticket_mensal = clientes_df['Valor Contrato Mensal'].mean()
print('Ticket Médio Mensal R$ {:,.2f}'.format(ticket_mensal))
```
